## Import libraries and get data

```python
import numpy as np
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

Generate sample data for clustering

```python
# Generate a synthetic dataset
data, _ = make_blobs(n_samples=300, centers=4, random_state=42)
```

Data visualization

```python
# Visualize raw data
plt.scatter(data[:, 0], data[:, 1])
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Raw Data')
plt.show()
```

Clustering
- Kmeans
- Density-Based Spatial Clustering
- AgglomerativeClustering (a type of hierarchical clustering algorithm)

## Initialize Kmeans clustering

Define
- number of clusters (only one hyper-paramter)

```python
# Create a K-Means model and fit it to the data
kmeans = KMeans(n_clusters=4, random_state=42)
```

fit the data 

```python
kmeans.fit(data)
```

Plot the samples into with cluster centers

```python
# Visualize the clustered data
plt.scatter(data[:, 0], data[:, 1], c=kmeans.labels_, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], marker='x', s=100, c='red')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('K-Means Clustering')
plt.show()
```

## Initialize AgglomerativeClustering
Define
- number of cluster centers

```python
from sklearn.cluster import AgglomerativeClustering

# Create an AgglomerativeClustering model and fit it to the data
agg_clustering = AgglomerativeClustering(n_clusters=4)
agg_labels = agg_clustering.fit_predict(data)

# Visualize the clustered data
plt.scatter(data[:, 0], data[:, 1], c=agg_labels, cmap='viridis')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Agglomerative Clustering')
plt.show()
```

## Initialize DBSCAN
Define
- eps
- min_samples

```python
from sklearn.cluster import DBSCAN

# Create a DBSCAN model and fit it to the data
dbscan = DBSCAN(eps=0.5, min_samples=5)
dbscan_labels = dbscan.fit_predict(data)

# Visualize the clustered data
plt.scatter(data[:, 0], data[:, 1], c=dbscan_labels, cmap='viridis')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('DBSCAN Clustering')
plt.show()
```
