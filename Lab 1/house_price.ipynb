Now, let's find the coorelation between features in this dataset

### 5. Correlations

```python
fig,ax = plt.subplots(figsize=(10,8))
correlation = df.select_dtypes(include=['float64','int64']).iloc[:,1:].corr()
sns.heatmap(correlation,ax=ax,vmax=1,square=True)
plt.title('Numeric Feature heatmap')
```

Now let's find the correlation between SalePrice and other features

```python
corr_dict = correlation['SalePrice'].to_dict()
del correlation['SalePrice']
for key,val in sorted(corr_dict.items(),key=lambda x:-abs(x[1])):
    print('{0} \t : {1}' .format(key,val))
```

correlation between SalePrice and OverallQual
- there is a positive correlation

```python
sale_price = df['SalePrice']
overall_qual = df['OverallQual']
plt.scatter(sale_price, overall_qual)
plt.plot([0, 400000], [0, 10], color='red', label='simple linear model')
plt.legend()
plt.title('Correlation between Sale Price and OverallQual')
plt.xlabel('Sale Price')
plt.ylabel('OverallQual')
plt.show()
```

correlation between SalePrice and 3SsnPorch
- there is no correlation

```python
sale_price = df['SalePrice']
overall_qual = df['3SsnPorch']
plt.scatter(sale_price, overall_qual)
# plt.plot([0, 400000], [0, 10], color='red', label='simple linear model')
plt.legend()
plt.title('Correlation between Sale Price and 3SsnPorch')
plt.xlabel('Sale Price')
plt.ylabel('3SsnPorch')
plt.show()
```
