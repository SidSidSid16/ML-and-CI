## Sample data

### 1. Import necessary python packages

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

### 2. Generate data

Sample data
- x: generate 100 samples between 0 to 10
- y: compute the labels using the given equation ==> 2*x + 1 + random(0, 100)

```python
np.random.seed(0)
x = np.linspace(0, 10, 100)
y = 2 * x + 1 + np.random.randn(100)
```

### 3. Plot data points

Scatter plot data on X-axis and labels on Y-axis

```python
plt.scatter(x, y, label='Data points')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot')
plt.legend()
plt.show()
```

Scatter plot shows that there is a linear relationship between features (x-axis) and labels (y-axis). 
- a linear model would be a good choice for this type of data
- at x=0, y=1 and a = 2. Thus a simple linear line equation y = ax + b ==> y = 2*x + 1
- depending on the distribution of data, the equation will change

### 4. Plot linear model and data points

```python
# a simple linear line
plt.scatter(x, y, label='Data points')
plt.plot([0, 10], [1, 21], color='red', label='Linear Line/Model')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot')
plt.legend()
plt.show()
```

### 5. Simple error estimation

Let's find error when predicting a label for a given feature

```python
test_feature = x[23]
test_label = y[23]

a = 2
b = 1 
predicted_label =  a * test_feature + b
```

```python
print('test_label: ', test_label)
print('predicted_label: ', predicted_label)
error = abs(test_label - predicted_label)

print('error: ', error)
```

### 8. Correlations

Correlations are very useful to identify any trends in the data

```python
# Create a sample dataset
data = {
    'Feature1': np.random.rand(100),
    'Feature2': np.random.rand(100),
    'Feature3': np.random.rand(100),
    'Feature4': np.random.rand(100)
}

df = pd.DataFrame(data)
df.head()
```

Correlation or dependence is any statistical relationship, whether causal or not, between two random variables or bivariate data. Coorelation types:
- positive correlation
- negative correlation
- no correlation

```python
display.Image("correlation.png")
```

In the figure below we will plot the first two features. By visual inspection, can you tell if Feature1 and Feature2 are correlated?

```python
# Visualize data with scatter plot
f1 = df['Feature1']
f2 = df['Feature2']
plt.scatter(f1, f2)
plt.xlabel('Feature1')
plt.ylabel('Feature2')
plt.title('Scatter Plot')
plt.show()
```

Pandas has inbuild methods to compute correlation between features.
Let's find the correlation between four features in the above sample data. 

```python
# Calculate the correlation matrix
correlation_matrix = df.corr()

# Create a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Heatmap')
plt.show()
```

From the heatmap can you tell if the there is any correlation between each feature?
